```python
#import basic libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
##Divide the dataset into independant and dependant features

X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
print(X, y)
```

```python
#Feature Engineering
geography = pd.get_dummies(X['Geography'], drop_first=True, dtype='int')
gender = pd.get_dummies(X['Gender'], drop_first=True, dtype = 'int')
```

```python
#Join it back to dataframe
X = X.drop(['Geography', 'Gender'], axis = 1)
X.head()
```

```python
X = pd.concat([X, geography, gender], axis = 1)
X
```

```python
#Splitting the dataset into training asnd test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler #based on z-score
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
#Next Part, let create the ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, ReLU, PReLU, ELU
from tensorflow.keras.layers import Dropout
```

```python
###Lets initialize the ANN
classifier  = Sequential()
```

```python
##Adding the input layer
classifier.add(Dense(units=11, activation='relu'))
```

```python
###Adding the first hidden layer
classifier.add(Dense(units=7, activation='relu'))
###Adding Dropout
#classifier.add(Dropout(0.2))
```

```python
###Adding the second hidden layer
classifier.add(Dense(units=6, activation='relu'))
###Adding Dropout
#classifier.add(Dropout(0.3))
```

```python
###Adding the output layer
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
###early stopping when the accuracy is not increasing the training of the model will stop automatically
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor = "val_loss",
    min_delta = 0.0001,
    patience = 20,
    verbose = 1,
    mode = "auto",
    baseline = None,
    restore_best_weights = False
)
```

```python
model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=1000, callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#Summarize history for accuracy
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#Part 3 - Making the predictions and evaluating the model

#Predicting the test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred>=0.5)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size = 5000
messages = X.copy()
messages.reset_index(inplace=True)
messages['title'][1]
```

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM, Dense, Bidirectional
```

```python
###Vocabulary Size
voc_size = 5000

messages = X.copy()
messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
messages.reset_index(inplace=True)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[0]
```

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr[0]
```

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen=sent_length)
embedded_docs[0]
```

```python
#Creating a model
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features))
model.build(input_shape=(None, sent_length))
model.add(Bidirectional(LSTM(100)))#number of neurons
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs), y.shape
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=32)
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=10, batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np

y_pred = np.where(y_pred>=0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred), "\n Accuracy-", accuracy_score(y_test, y_pred))
```
